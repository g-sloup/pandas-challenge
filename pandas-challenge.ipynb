```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

```python
purchase_data_df.head()
```

## Player Count

```python
# create a data frame with new column named Total Players with count of unique values for SN column
SN_count = len(purchase_data_df["SN"].unique())

# Place all of the data found into a summary DataFrame
totalplayer_df = pd.DataFrame({"Total Players": [SN_count]})

totalplayer_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate the number of unique items in the DataFrame
unique_items = len(purchase_data_df["Item ID"].unique())

# Calculate average price of items - mean
ave_price = (purchase_data_df["Price"]).mean()

# Calculate total number of purchases - len
total_purchases = len(purchase_data_df["Purchase ID"])

# Calculate total revenue  - sum
total_revenue = sum(purchase_data_df['Price'])

# Place all of the data found into a summary DataFrame
summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Price": ave_price,
                              "Number of Purchases": total_purchases,
                              "Total Revenue": total_revenue})

summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

summary_df
```

## Gender Demographics

```python
# Drop duplicates on SN
nodupesSN_df = purchase_data_df.drop_duplicates(subset=['SN'])

nodupesSN_df.head()
```

```python
gender_count = nodupesSN_df["Gender"].value_counts()

gender_percent = nodupesSN_df['Gender'].value_counts(normalize=True) * 100

gender_df = pd.DataFrame({"Total Count": gender_count,
                              "Percent of Total Players": gender_percent})

gender_df["Percent of Total Players"] = gender_df["Percent of Total Players"].map("{:.2f}%".format)

gender_df
```


## Purchasing Analysis (Gender)

```python
purch_count = purchase_data_df["Gender"].value_counts()

ave_purch = purchase_data_df.groupby("Gender").mean()["Price"]

total_value = purchase_data_df.groupby("Gender").sum()["Price"]

per_person = total_value / gender_count

gender_purch_df = pd.DataFrame({"Purchase Count": purch_count, "Average Purchase Price": ave_purch, 
                                "Total Purchase Value": total_value,
                                "Average Total Per Person": per_person,
                               })


gender_purch_df["Average Purchase Price"] = gender_purch_df["Average Purchase Price"].map("${:.2f}".format)

gender_purch_df["Total Purchase Value"] = gender_purch_df["Total Purchase Value"].map("${:,.2f}".format)

gender_purch_df["Average Total Per Person"] = gender_purch_df["Average Total Per Person"].map("${:.2f}".format)

gender_purch_df
```

## Age Demographics

```python
# Determine max age for bins
max_age = purchase_data_df["Age"].max()
max_age
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
group_names
```

```python
# Add Age Range column with bins to original dataframe
purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names, include_lowest=True)
purchase_data_df.head()
```

```python
# Create a new dataframe that drops the duplicates, after age range column added. 
# This was done because a Warning was appearing when running pd.cut to add Age Range on the nodupes_SN dataframe. 

nodupes_df = purchase_data_df.drop_duplicates(subset=['SN'])
nodupes_df.shape
```

```python
age_count = nodupes_df["Age Range"].value_counts()

age_percent = nodupes_df['Age Range'].value_counts(normalize=True,sort=False) * 100

age_df = pd.DataFrame({"Total Count": age_count,
                              "Percent of Total Players": age_percent})

age_df["Percent of Total Players"] = age_df["Percent of Total Players"].map("{:.2f}%".format)

age_df
```

## Purchasing Analysis (Age)

```python
purch_count = purchase_data_df["Age Range"].value_counts()

ave_purch = purchase_data_df.groupby("Age Range").mean()["Price"]

total_value = purchase_data_df.groupby("Age Range").sum()["Price"]

per_person = total_value / age_count

age_purch_df = pd.DataFrame({"Purchase Count": purch_count, "Average Purchase Price": ave_purch, 
                                "Total Purchase Value": total_value,
                                "Average Total Per Person": per_person,
                               })

age_purch_df.index.name = "Age Range"

age_purch_df["Average Purchase Price"] = age_purch_df["Average Purchase Price"].map("${:.2f}".format)

age_purch_df["Total Purchase Value"] = age_purch_df["Total Purchase Value"].map("${:,.2f}".format)

age_purch_df["Average Total Per Person"] = age_purch_df["Average Total Per Person"].map("${:.2f}".format)

age_purch_df
```

## Top Spenders

```python
purchase_count = purchase_data_df["SN"].value_counts()

ave_purch = purchase_data_df.groupby("SN").mean()["Price"]

total_value = purchase_data_df.groupby("SN").sum()["Price"]

SN_purch_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": ave_purch, 
                                "Total Purchase Value": total_value,
                               }).sort_values("Total Purchase Value", ascending=False)

SN_purch_df["Average Purchase Price"] = SN_purch_df["Average Purchase Price"].map("${:.2f}".format)

SN_purch_df["Total Purchase Value"] = SN_purch_df["Total Purchase Value"].map("${:,.2f}".format)


SN_purch_df.head()
```

## Most Popular Items

```python
item_price_df = purchase_data_df[["Item ID","Item Name","Price"]]

item_price_df.head()
```

```python
nodupesItem_df = item_price_df.drop_duplicates(subset=['Item ID'])

nodupesItem_df.head()
```

```python
nodupesItem_df["Total Purchase Value"] = item_price_df.groupby("Item ID").sum()["Price"]
nodupesItem_df["Purchase Count"] = item_price_df["Item ID"].value_counts()
nodupesItem_df.sort_values("Purchase Count", ascending=False)
nodupesItem_df.head()
```

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame


```python
purchase_count = item_price_df["Item ID"].value_counts()

total_value = item_price_df.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

item_price = total_value / purchase_count

popular_item_df = pd.DataFrame({"Purchase Count": purchase_count, "Item Price": item_price,
                                "Total Purchase Value": total_value
                               }).sort_values("Purchase Count", ascending=False)

popular_item_df.index.name = "Item ID"

#popular_item_df["Item Price"] = popular_item_df["Item Price"].map("${:.2f}".format)
#popular_item_df["Total Purchase Value"] = popular_item_df["Total Purchase Value"].map("${:.2f}".format)

popular_item_df.head()
```
